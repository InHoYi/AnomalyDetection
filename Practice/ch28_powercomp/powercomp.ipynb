## 상관관계

0.0 <= |r| < 0.2 : 상관관계가 없다, 선형의 관계가 없다.

0.2 <= |r| < 0.4 : 약한 상관관계가 있다.

0.4 <= |r| < 0.6 : 보통의 상관관계가 있다.

0.6 <= |r| < 0.8 : 강한 상관관계가 있다.

0.8 <= |r| < 1.0 : 매우 강한 상관관계가 있다.

```python
X_train = raw.drop(['DateTime', 'zone1', 'zone2', 'zone3'], axis = 1)
X_train.head()
```

```python
from sklearn.ensemble import IsolationForest
# n_estimators : 나무의 개수
# max_samples : 샘플링 데이터 개수 (정수값을 넣으면 원하는 개수 지목, 0 ~ 1 값을 넣으면 비율을 선택)
# contaminaton : 이상치의 비율 설정 (score의 threshold를 정의)
# max_features : 학습 시 사용할 속성의 개수
```

```python
clf = IsolationForest(max_samples = 0.3, contamination = 0.10, random_state = 42)
clf.fit(X_train)
y_pred_train = clf.predict(X_train)
```

```python
y_pred_train = np.where(y_pred_train == -1, 1, 0)
```

```python
raw['abnormal'] = y_pred_train
```

```python
raw['abnormal'].value_counts()
```

```python
raw_split = raw[raw['DateTime'].astype(str).str.contains('2017-08')]
plt.figure(figsize = (25,6))
plt.plot(raw_split[raw_split['abnormal'] == 1]['DateTime'], raw_split[raw_split['abnormal'] == 1]['zone1'], 'o', color = 'red', markersize = 5)
plt.plot(raw_split['DateTime'], raw_split['zone1'], linestyle = '--', color = 'grey')
```

```python
import numpy as np
import pandas as pd
import datetime
```

```python
raw = pd.read_csv('chapter28_powercomp.csv')
raw['DateTime'] = pd.to_datetime(raw['DateTime'])
raw.sort_values(by = ['DateTime'], ascending = True)
```

```python
print(raw.isnull().sum())
```

```python
raw.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('dark_background')
plt.figure(figsize=(25,6))
plt.plot(raw['DateTime'], raw['Zone 1 Power Consumption'], label = 'zone1')
plt.plot(raw['DateTime'], raw['Zone 2  Power Consumption'], label = 'zone2')
plt.plot(raw['DateTime'], raw['Zone 3  Power Consumption'], label = 'zone3')
plt.ylabel('Power Comsumption')
plt.xlabel('DateTime')
plt.legend()
```

```python
fig, ax1 = plt.subplots(figsize = (25,6))

ax1.plot(raw['DateTime'], raw['Zone 1 Power Consumption'], color = 'green', label = 'zone1')

ax2 = ax1.twinx()
ax2.plot(raw['DateTime'], raw['diffuse flows'], color = 'deeppink', label = 'Diffuse')

fig.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize = (25, 6))
# Zone 1 Power Consumption, genral diffuse flows

ax1.plot(raw['DateTime'], raw['Zone 1 Power Consumption'], color = 'green', label = 'zone1')

ax2 = ax1.twinx()
ax2.plot(raw['DateTime'], raw['general diffuse flows'], color = 'blue', label = 'general')

fig.legend()
fig.show()
```

```python
fig, ax1 = plt.subplots(figsize = (25, 6))
# Zone 1 Power Consumption, genral diffuse flows

ax1.plot(raw['DateTime'], raw['Zone 1 Power Consumption'], color = 'green', label = 'zone1')

ax2 = ax1.twinx()
ax2.plot(raw['DateTime'], raw['Temperature'], color = 'purple', label = 'temp')

fig.legend()
fig.show()
```

```python
raw['hour'] = raw['DateTime'].dt.hour
```

```python
zone1_by_hour = raw['zone1'].groupby(raw['hour']).mean().sort_values(ascending = False).reset_index()
zone1_by_hour.sort_values('hour', ascending = True)

plt.bar(zone1_by_hour['hour'], zone1_by_hour['zone1'])
plt.show()
```
